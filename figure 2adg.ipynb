```python
import mne
import os
import pandas as pd
import numpy as np
from scipy import signal, stats
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_excel(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure2adg_data.xlsx',
                       index_col = [0])
```

```python
freqs = np.array(data_df['freqs'])
```

```python
dys_wake_mean = np.array(data_df['dys_wake_mean'])
dys_wake_max = np.array(data_df['dys_wake_max'])
dys_wake_min = np.array(data_df['dys_wake_min'])
dys_rem_mean = np.array(data_df['dys_rem_mean'])
dys_rem_max = np.array(data_df['dys_rem_max'])
dys_rem_min = np.array(data_df['dys_rem_min'])
dys_nrem_mean = np.array(data_df['dys_nrem_mean'])
dys_nrem_max = np.array(data_df['dys_nrem_max'])
dys_nrem_min = np.array(data_df['dys_nrem_min'])
```

```python
pds_wake_mean = np.array(data_df['pds_wake_mean'])
pds_wake_max = np.array(data_df['pds_wake_max'])
pds_wake_min = np.array(data_df['pds_wake_min'])
pds_rem_mean = np.array(data_df['pds_rem_mean'])
pds_rem_max = np.array(data_df['pds_rem_max'])
pds_rem_min = np.array(data_df['pds_rem_min'])
pds_nrem_mean = np.array(data_df['pds_nrem_mean'])
pds_nrem_max = np.array(data_df['pds_nrem_max'])
pds_nrem_min = np.array(data_df['pds_nrem_min'])
```

```python
fig, ax = plt.subplots(figsize=(4,3))
plt.rcParams.update({'font.size': 12})
#peace_green
line1, = ax.plot(freqs, dys_wake_mean, color = 'darkred')
ax.plot(freqs, dys_wake_min, color = 'gray',alpha=0.1)
ax.plot(freqs, dys_wake_max, color = 'gray' ,alpha=0.1)
ax.fill_between(freqs, dys_wake_min, dys_wake_max, color = 'darkred',alpha=0.5)

line2, = ax.plot(freqs, pds_wake_mean, color = 'darkorange')
plt.axvspan(6, 11.5, color='lightblue', alpha=0.3, lw=0)
plt.axvspan(13.5, 30, color='lightblue', alpha=0.3, lw=0)
ax.plot(freqs, pds_wake_min, color = 'gray', alpha=0.1)
ax.plot(freqs, pds_wake_max, color = 'gray' ,alpha=0.1)
ax.fill_between(freqs, pds_wake_min, pds_wake_max, color = 'darkorange',alpha=0.5)

ax.legend([line1, line2,], ['DYS', 'PD'], fancybox = True, )
ax.set_ylim(0, 0.06)
ax.set_xlim(5, 33)
ax.set_xlabel('Frequency (Hz)',fontsize=13)
ax.set_ylabel('Rel.spectral power (%)',fontsize=14)
ax.set_title('PD vs. DYS for Awake state',fontsize=14)
plt.tight_layout()
```
