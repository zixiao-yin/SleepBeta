```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
mean_nrem = np.load(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure4f_nremmean.npy')
error_nrem = np.load(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure4f_nremerror.npy')
data_nrem = pd.read_excel(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure4f_nremdata.xlsx',
                         index_col = [0])
```

```python
sns.set_context("paper", font_scale=1.5)
fig, ax = plt.subplots(figsize=(6 ,3))
x_data = [0,1,2,3]
ax.errorbar(x_data, mean_nrem, error_nrem, linewidth =3, color = 'lightblue')
sns.stripplot(x='name' , y='data_all', data=data_nrem,   order = ['fir', 'sec', 'thi', 'fot'],
              palette=['darkblue', 'darkred', 'darkorange','darkgreen'],size = 7,alpha=.7)

ax.plot(x_data, mean_nrem, '-o', color = 'lightblue',markersize=8)
ax.set(xticks = [ 0 ,1 ,2 ,3], xticklabels = ['First', 'Second', 'Third', 'Fourth'])
ax.set(ylim = (0,0.7))
ax.set_xlabel('Sleep duration quantiles',fontsize=13)
ax.set_ylabel('Beta power (%)',fontsize=14)
ax.set_title('Beta power in NREM sleep',fontsize=14)
plt.tight_layout()
```
