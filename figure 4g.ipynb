```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
nrem_mean =  np.load(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure4g_nremmean.npy')
nrem_error =  np.load(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure4g_nremerror.npy')
nrem_data = pd.read_excel(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure4g_nremdata.xlsx',
                         index_col = [0])
```

```python
sns.set_context("paper", font_scale=1.5)
fig, ax = plt.subplots(figsize=(6 ,3))
x_data = [0,1,2,3]
ax.errorbar(x_data, nrem_mean, nrem_error, linewidth =3, color = 'lightblue')
sns.stripplot(x='name' , y='nrem_all', data=nrem_data,   order = ['fir', 'sec', 'thi', 'fot'],
              palette=['darkblue', 'darkred', 'darkorange','darkgreen'],size = 7,alpha=.7)

ax.plot(x_data, nrem_mean, '-o', color = 'lightblue',markersize=8)
ax.set(xticks = [ 0 ,1 ,2 ,3], xticklabels = ['First', 'Second', 'Third', 'Fourth'])
ax.set(ylim = (-5,105), yticks = [ 0, 25, 50, 75,100])
ax.set_xlabel('Sleep duration quantiles',fontsize=13)
ax.set_ylabel('Time percentage (%)',fontsize=14)
ax.set_title('Stimulation triggered in NREM sleep',fontsize=14)
plt.tight_layout()
```
