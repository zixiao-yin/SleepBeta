```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
```

```python
pd_sub = pd.read_excel(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure3ab_pdsub.xlsx',
                      index_col = [0])
pd_rem = pd_sub[pd_sub['Beta-REM'].notna()]
dys_sub = pd.read_excel(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure3ab_dyssub.xlsx',
                      index_col = [0])
```

```python
sns.set_context("paper", font_scale=1.2)
fig, ax = plt.subplots(figsize=(3.5 ,3))
line1 = sns.regplot(x='Beta-NREM', y='PSQI', data=pd_sub, color = 'darkred',
                    label = 'PD')
line2 = sns.regplot(x='Beta-NREM', y='PSQI', data=dys_sub, color = 'gray',
                  label = 'DYS')

ax.legend(['PD (rho = 0.63, $\it{P}$ = 0.03)', 'DYS (rho = 0.19, $\it{P}$ = 0.43)',], 
          fancybox = True, prop={'size': 11} )
#ax.set(xlim = (0.01, 0.3), xticks = [ 0.1,0.2,])
ax.set_xlabel('Beta power in NREM sleep (%)')
ax.set_ylabel('PSQI')
ax.set_title('NREM Beta-PSQI')
plt.tight_layout()
```

```python
sns.set_context("paper", font_scale=1.2)
fig, ax = plt.subplots(figsize=(3.5 ,3))
sns.regplot(x='Beta-NREM', y='RBDSQ', data=pd_sub, color = 'darkorange',label = 'NREM')
sns.regplot(x='Beta-REM', y='RBDSQ', data=pd_rem, color = 'gray',label = 'REM')
ax.legend( ['NREM (rho = 0.67, $\it{P}$ = 0.02)', 'REM (rho = 0.55, $\it{P}$ = 0.10)',], 
          fancybox = True,prop={'size': 11} )
#ax.set(xlim = (0.01, 0.3), xticks = [ 0.1,0.2,])
ax.set_xlabel('Beta power during sleep (%)')
ax.set_ylabel('RBDSQ')
ax.set_title('NREM/REM Beta - RBDSQ')
plt.tight_layout()
```
