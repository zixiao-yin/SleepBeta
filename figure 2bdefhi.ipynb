```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel(r'C:\Users\Lenovo\1代码 REM中beta升高\upload_data\figure2bdefhi_data.xlsx')
```

```python
nrem_all = df.groupby('state').get_group('nrem').reset_index(drop = True)
rem_all = df.groupby('state').get_group('rem').reset_index(drop = True)
wake_all = df.groupby('state').get_group('restWake').reset_index(drop = True)
```

```python
x = 'disease'
y = 'psdAllBeta'
data = nrem_all
sns.set_context("paper", font_scale=1.4)
fig, ax = plt.subplots(figsize=(2.5 ,3))
box_plot = sns.boxplot(x=x , y=y, data=data, linewidth=2, saturation =1,
                        order = ['dys', 'pd'], color = 'skyblue',  width = 0.4,
                       boxprops=dict(alpha=.3))
sns.stripplot(x=x , y=y, data=data,   order = ['dys', 'pd'],
              palette=['darkred', 'darkorange'],size = 7,alpha=.7)
box_plot.set(ylim = (0,0.8), yticks = [ 0.1, 0.3,0.5, 0.7])
box_plot.set_xticklabels ([ 'DYS', 'PD'])
box_plot.set(xlabel='', ylabel= 'Beta power (%)', title = 'NREM Beta')
plt.tight_layout()
```
